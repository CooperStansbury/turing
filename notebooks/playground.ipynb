```python
import numpy as np
from scipy.integrate import odeint
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def uncoupledModel(x, t, a):
    x_max = 1
    dxdt = a * (x - x_max)
    return dxdt

a = -0.5
x0 = np.linspace(0,1,20)t
t = np.linspace(0,10,400)
res = odeint(uncoupledModel,x0,t,args=(a,))

res = pd.DataFrame(res, columns = x0)
res = res.reset_index(drop=False, names='t')

res = pd.melt(res, id_vars='t')
res.columns = ['t', 'init', 'value']


plt.rcParams['figure.dpi'] = 300
sns.lineplot(data=res, 
             x='t',
             y='value', 
             hue='init')


sns.despine()
plt.ylabel("x")
plt.xlabel("t")

```
